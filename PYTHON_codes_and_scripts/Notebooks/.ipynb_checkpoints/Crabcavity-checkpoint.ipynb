```python
import pyoptics as opt

beam='1'

HLLHC_60cm='/afs/cern.ch/user/n/nbiancac/ln_work/scratch0/IRIS/Getting_LHC_beta_functions/2015/'+\
'HLLHC-squeeze60cm/twiss_lhcb'+beam+'.tfs'
HLLHC_15cm='/afs/cern.ch/user/n/nbiancac/ln_work/scratch0/IRIS/Getting_LHC_beta_functions/2015/'+\
'HLLHCV1.1/twiss_lhcb'+beam+'.tfs'

mad_optics=opt.optics.open(HLLHC_15cm)

dire='/afs/cern.ch/user/n/nbiancac/ln_work/scratch0/IRIS/PYTHON_codes_and_scripts/LHC_impedance_and_scripts/LHC_elements/HOMs/'
CCtypesIP=['DQW_20151001','DQW_20151001']
filenames=[dire+'HLLHC_Crab_cavities_'+CCtype+'.txt' for CCtype in CCtypesIP]

Ncav=4 # CC/IP

betavy=machine.R/machine.Qy
betavx=machine.R/machine.Qx

imp_mod=[]; wake_mod=[];
for iIPn,IPn in enumerate(['1','5']):
    for ii in np.arange(Ncav):
        # optics
#         mad_optics.show('ACF[ABC].*('+IPn+'.B'+beam+'|'+IPn+'B'+beam+')','s betx bety')
        idx=mad_optics.pattern('ACF[ABC].*('+IPn+'.B'+beam+'|'+IPn+'B'+beam+')')
        betx_CC=[];bety_CC=[];
        for name in mad_optics.name[idx]:
            bety_CC.append(mad_optics.twissdata(name,'bety')['bety'])
            betx_CC.append(mad_optics.twissdata(name,'betx')['betx'])

        betx_CC_av=np.mean(betx_CC)
        bety_CC_av=np.mean(bety_CC)
        print 'Average betax and betay at IP'+IPn+':%.1f, %.1f'%(betx_CC_av,  bety_CC_av)

        # HOMs impedance
        filename=filenames[iIP];
        imp,wake=imp_model_from_HOMfile(filename,unit_freq='Hz',spread=True, spread_rng=3e6, seed=iIPn*ii, 
                                        damp_fm=True, beta=1,fpar=freq_param(ftypescan=0,nflog=100),zpar=z_param())

        if IPn=='5':
            imp=rotate_imp_wake(imp,pi/2)
            wake=rotate_imp_wake(wake,pi/2)

        # weight
        add_impedance_wake(imp_mod, imp, betx_CC_av/betavx, bety_CC_av/betavy)
        add_impedance_wake(wake_mod,wake,betx_CC_av/betavx, bety_CC_av/betavy)
    


a,b,c,d,plane,wakeflag=identify_component('Zydip')
for iw in imp_mod:
    if test_impedance_wake_comp(iw, a, b, c, d, plane):
        Zy=iw.func[:,0]+1j*iw.func[:,1]
a,b,c,d,plane,wakeflag=identify_component('Zxdip')
for iw in imp_mod:
    if test_impedance_wake_comp(iw, a, b, c, d, plane):
        Zx=iw.func[:,0]+1j*iw.func[:,1]
        freq=iw.var;       
        

figure(1)
semilogy(freq,real(Zx))
semilogy(freq,real(Zy),'-r')
xlim(0, 5e9)
ylim(1e3,1e10)
```
