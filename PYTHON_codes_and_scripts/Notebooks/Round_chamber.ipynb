# SLS-II 
Their copper vacuum chamber has rounded geometry with inner radius of 10 mm and a thickness of 
NEG coating is, say 2um.  They are using a DC resistivity for NEG of 9.1e-7 Ohm*m and 
for copper of 1.68e-8 Ohm*m. The thickness of the copper chamber is 1mm.

We calculated the longitudinal impedance up to 100THz and the transverse impedance up to 
1THz for comparison with Impedanacewake2D.

```python
%pylab notebook
from string import *
from copy import deepcopy
import os,re
path_here=os.getcwd()+"/";
from particle_param import *
from Impedance import *
import itertools
```

```python
Rin=10e-3;
neg_lay=Z_layer(rhoDC=9.1e-7,thickness=2e-6)

cu_lay=Z_layer(rhoDC=1.68e-8,thickness=1e-3)
vac_lay=vacuum_layer(thickness=np.inf)

pipe=construct_layers([neg_lay,cu_lay,vac_lay])
freq=freq_param(fmin=1e2,fmax=1e14,ftypescan=2,nflog=20)
iw_input=impedance_wake_input(machine='SLSII',b=[Rin],layers=pipe,fpar=freq,geometry='round')

namefile='slsII.txt';
write_impedance_wake_input(namefile,iw_input)
# os.system('roundchamber.x < slsII.txt')
os.system('wake_roundchamber.x < slsII.txt')


ResultDir='/afs/cern.ch/user/n/nbiancac/scratch0/IMPEDANCE/SLSII/'
os.system('mv *.txt W*dat Z*dat Input*.dat '+ResultDir)
%cd $ResultDir
```
